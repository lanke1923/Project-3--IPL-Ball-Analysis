Now we shall analyze the data to fin dout the proportion of winnings by each team every year.

```python
#using the for loop to select each year from the list of the years games were played
for i in df.year.unique():
    print('Year :',i,'<','='*50,'>')
    xdf=df[df.year==i]
    plt.pie(xdf.winner.value_counts(),autopct="%1.1f%%")
    plt.legend(xdf.winner.value_counts().index,bbox_to_anchor=(2,1),loc='upper right',title='Teams in {}'.format(i))
    plt.show();
```
